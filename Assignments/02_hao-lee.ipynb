```python
import geopandas as gpd
import os
```

## Creating the geojson

```python
kml_file = 'doc.kml'
geojson_path = 'taichung.geojson'

gdf = gpd.read_file(kml_file, driver='KML')

gdf.to_file(geojson_path, driver='GeoJSON')
print(f"GeoJSON file saved to: {geojson_path}")
```

## Central District, Taichung, Taiwan's old theaters (built in the 1970s)

```python
import matplotlib.pyplot as plt
import contextily as ctx

# Read the GeoJSON file
gdf = gpd.read_file(geojson_path)

# Plot the GeoDataFrame with a background map
fig, ax = plt.subplots(1, 1, figsize=(10, 10))
gdf.plot(ax=ax, color='black', edgecolor='black')
ax.set_axis_off()

# Add greyscale OSM basemap without icons
ctx.add_basemap(ax, crs=gdf.crs.to_string(), source=ctx.providers.OpenStreetMap.HOT, alpha=0.5)

plt.show()
```
